```python

from gensim.models.keyedvectors import KeyedVectors
glove = KeyedVectors.load_word2vec_format("load/glove.6B.300d.txt.w2v", binary=False)
```

```python
with open("load/stopwords.txt", 'r') as r:
    stops = []
    for line in r:
        stops += [i.strip() for i in line.split('\t')]
```

```python

u=len(stops)
for x in range(u):
    word=stops[x]
    wor= word[0].upper()
    for i in range(1,len(word)):
        wor+=word[i]
    stops.append(wor)
print(stops)
```

```python
def su (text,gram,hyp):
    lookup = [] #"Words to look up"
    assert gram >0, "invalid gram size"
    assert len(text.split())>=gram, "Your Gram size cant be greater than your word input"
    text = re.compile('[{}]'.format(re.escape(string.punctuation))).sub(" ",text)
    dic = {}
    l=[]
    Tex=text.lower().split()
    for p in stops:
        if p in l:
            l.remove(p)
    X=0    
    for i in range(len(Tex)):
        
        if i+gram== len(Tex):
            break
        else:
            phrase=""
            for k in range(gram):
                phrase+=Tex[i+k] + " "
            l.append(phrase)
            for m in l:
                words=m.split()
                sem = np.zeros((1,300))
                try:
                    for k in words:
                        sem+=glove[k]
                except Exception as ex:
                    print(ex)
                    lookup.append(k)
                    dic.update({m:sem})
        print(f"This is the {X}th iteration")
    for i in range(len(l)):
        for y in range(i,len(l)):
            if np.dot(dic[l[i]],dic[l[y]]) ==hyp:
                l.remove(y)
    print(l)
    return l
```
